# nlmrs: Neutral Landscape Models in Python

This notebook contains examples for all algorithms available in the `nlmrs` library.

## Setup
Install the library and dependencies.

```python
# Install nlmrs from PyPI
!pip install nlmrs matplotlib numpy

import nlmrs
import matplotlib.pyplot as plt
import numpy as np

def plot_grid(grid, title):
    plt.figure(figsize=(6, 6))
    plt.imshow(grid, cmap='terrain', vmin=0, vmax=1)
    plt.title(title)
    plt.axis('off')
    plt.show()
```

### Random

```python
grid = nlmrs.random(100, 100, **{}, seed=42)
plot_grid(grid, 'Random')
```
